```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12,34,25,56,62,69,78,89,98,100])
y = np.array([8,9,11,16,34,38,45,56,67,78])
```

```python
# Line plot
plt.title('Line Plot')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.grid()
plt.plot(x, y)
```

```python
t = np.arange(0,101,10)
u = np.arange(20,31)
```

```python
t
```

```python
# Line plot
plt.title('Line Plot')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.grid()
plt.plot(t, u)
```

```python
# Line plot
plt.title('Line Plot')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.xlim(0, 110)
plt.ylim(0, 80)
plt.xticks(range(0,110,10))
plt.grid()
plt.plot(x, y, color='red', marker='o', ls='-.', ms=10, mfc='green')
# markers: . o < > ^ v D 
# ls- : --
```

```python
z = np.array([23,21,26,28,35,47,56,78,94,95])
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.plot(x, y, color = 'r', label = 'X vs Y')
plt.plot(x, z, color = 'g', label = 'X vs Z')
plt.legend(loc='lower right')
```

```python
# Subplot
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('First Plot')
plt.grid()
plt.plot(x, y)
plt.subplot(1,2,2)
plt.title('Second Plot')
plt.grid()
plt.plot(x,z)
```

```python
w = np.array([56,48,45,43,42,36,21,18,16,5])
plt.plot(x, w)
```

```python
# Scatter plot
plt.title('Scatter Plot')
plt.scatter(x, y)
```

```python
# Bar plot
studs = [56,78,123,167,145,39,45]
langs = ['C','C++','Java','Python','PHP','JavaScript','R']
```

```python
plt.title('My Bar Plot')
plt.xlabel('Courses')
plt.ylabel('Students Enrolled')
plt.grid()
plt.bar(langs, studs, color = 'maroon')
```

```python
# Pie chart
plt.figure(figsize=(16,9))
plt.pie(studs, labels=langs, autopct='%2.2f%%', shadow=True);
```

```python
# Histgram
num = np.array([56,53,12,54,36,53,78,87,61,52,51,59,67,54,32,89])
```

```python
plt.title('Histogram')
plt.xlabel('Ranges')
plt.ylabel('Count')
plt.grid()
plt.ylim(0,10)
plt.xticks(range(0,101,10))
plt.hist(num, bins=range(0,101,10));
```

```python
# Box plot
plt.boxplot(num);
```
